## Importing Dependencies

```python
import pandas as pd
import numpy as np
import geocoder
```

## Reading Files

```python
data = pd.read_csv("toronto.csv")
data.head()
```

## Getting latitudes and longitudes of all the postal codes

```python
def get_lati_long(postal_code):
    lati_long = None
    while(lati_long is None):
        geo = geocoder.arcgis('{}, Toronto, Ontario'.format(postal_code))
        lati_long = geo.latlng
    return lati_long
```

```python
pcs = data['PostalCode']    
coordinates = [ get_lati_long(postal_code) for postal_code  in pcs.tolist() ]
```

```python
coordinates_df = pd.DataFrame(coordinates, columns=['Latitude', 'Longitude'])
data['Latitude'] = coordinates_df['Latitude']
data['Longitude'] = coordinates_df['Longitude']
```
